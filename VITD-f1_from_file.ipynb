```python
from pathlib import Path
import pandas as pd
from sklearn.metrics import f1_score
```

```python
PREDICTED_PATH = Path("/content/drive/MyDrive/research/VITD/predictions/mv5-71.866_71.808_71.977.csv")
TEST_PATH = Path("/content/drive/MyDrive/research/VITD/datasets/test.csv")
```

```python
# Load the actual labels from test.csv
predicted_df = pd.read_csv(PREDICTED_PATH)
predicted_labels = predicted_df["label"]
```

```python
# Load the predicted labels from test_predicted.csv
actual_df = pd.read_csv(TEST_PATH)
actual_labels = actual_df['label']
```

```python
# Calculate the F1-score
f1 = f1_score(actual_labels, predicted_labels, average='macro')

print(f'Macro F1-Score: {f1:.5f}')
```
